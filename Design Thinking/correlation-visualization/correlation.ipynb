```python
# ==============================
# Correlation Visualization Assignment
# ==============================

# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 2. Create Dataset
data = {
    'Maths': [85, 78, 92, 70, 88, 76, 95, 89],
    'Physics': [80, 75, 90, 65, 85, 70, 93, 87],
    'Chemistry': [78, 72, 88, 60, 82, 68, 91, 85],
    'English': [70, 65, 75, 55, 72, 60, 80, 78]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

# 3. Data Cleaning
print("\nMissing Values:\n", df.isnull().sum())
df = df.drop_duplicates().reset_index(drop=True)

# 4. Correlation Matrix
corr = df.corr()

print("\nCorrelation Matrix:")
print(corr)

# 5. Correlation Visualization (Improved Heatmap)
plt.figure()

plt.imshow(corr, cmap='coolwarm')   # better color
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)

# Add values inside heatmap
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        plt.text(j, i, round(corr.iloc[i, j], 2),
                 ha='center', va='center')

plt.title("Correlation Heatmap")

plt.show()

# 6. Interpretation
print("\nInterpretation:")
print("Values close to +1 → Strong positive correlation")
print("Values close to -1 → Strong negative correlation")
print("Values close to 0 → Weak or no correlation")
```
